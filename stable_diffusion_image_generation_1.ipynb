```python
# Step 3: Import libraries and load Stable Diffusion model

from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

# Move model to GPU
pipe = pipe.to("cuda")
```

```python
# Step 4: Generate an image from a text prompt

prompt = "A beautiful sunset over mountains with a lake"

image = pipe(prompt).images[0]

# Show the image
image
```

```python
# Step 5: Generate multiple images

prompt = "A futuristic city with flying cars at sunset"

images = pipe(prompt, num_images_per_prompt=3).images

# Display all generated images
for img in images:
    display(img)
```

```python
# Step 6: Save the generated image

prompt = "A futuristic city with flying cars at sunset"

image = pipe(prompt).images[0]

# Save the image
image.save("generated_image.png")

# Display the image
image
```
